## Non Shoping intent Query Exercise
For Non Shoping intented query identified the events and climate information and convert it into the Shopping Query


```python
import requests
import json
from datetime import date
import datetime as DT
import random
from IPython.display import display, Markdown
import openai

querystring = {
    "q": "United States",
    "appid" : "3b5b4c234a0970e4399f17179790eee8",
    "units": "imperial"
}

r = json.loads(requests.get('https://api.openweathermap.org/data/2.5/weather', params=querystring).text)

weather= r['weather'][0]['description']
temperature = r['main']['temp']

windSpeed=str(r['wind']['speed'])+" m/s"
if("rain" in r.keys()):
    Season="Rain"
elif(float(temperature)>70):
    Season="hot"
elif(float(temperature)<50):
    Season="cold"
else:
    Season=""

startday = date.today()
endday = date.today()+ DT.timedelta(days=7)
#print(startday,endday)


url = "https://api.getfestivo.com/v2/holidays"

querystring = {
	"api_key": "1b9ad697be61f1e849b706e5e1c35aa8",
	"country": "US",
	"year": 2021,
       'public': "true",
        "month":[startday.strftime("%m"),endday.strftime("%m")],
        "day":[startday.strftime("%d"),endday.strftime("%d")],
        "between":"true"
}

response = requests.request("GET", url, params=querystring).text
parsed = json.loads(response)
holidays=[]
for i in range(len(parsed["holidays"])):
    #print(parsed["holidays"][i]['name'])
    holidays.append(parsed["holidays"][i]['name'])
val=random.randint(0,len(holidays))
holiday=holidays[val-1]

gprompt='Create a attractive sentence for customers for '+Season+" Weather and "+holiday+' time sales of cloths.'
openai.api_key="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"

response = openai.Completion.create( engine="text-davinci-002", prompt=gprompt,  temperature=0.7, max_tokens=256, top_p=1.00, )
print(response['choices'][0]['text'])
products_Cat=[]
f = open('categories.json')
data = json.load(f)
f.close()
if Season == "hot":
    for i in range(len(data["Summer"])):
        
        products_Cat.append(data["Summer"][i])
       

elif Season == "cold":
    for i in range(len(data["Winter"])):
        products_Cat.append(data["Winter"][i])
elif Season == "Rain":
    for i in range(len(data["Rain"])):
        products_Cat.append(data["Rain"][i])
if(holiday in data.keys()):
    for i in range(len(data[holiday])):
        products_Cat.append(data[holiday][i])
print(products_Cat)
```
